```python
class Region(NamedTuple):

    id: str
    start: int
    end: int
    length: int
    strand: Literal["+", "-"] = "-"
    name: str | None = None

    def update(
        self,
        id: str | None = None,
        start: int | None = None,
        end: int | None = None,
        length: int | None = None,
        strand: Literal["+", "-"] | None = None,
        name: str | None = None,
    ):
        return self.__class__(
            self.id if id is None else id,
            self.start if start is None else start,
            self.end if end is None else end,
            self.length if length is None else length,
            self.strand if strand is None else strand,
            self.name if name is None else name
        )
```

```python
class Alignment(object):

    def __init__(
        self,
        ref_id: str,
        ref_start: int,
        ref_end: int,
        ref_length: int,
        ref_strand: str,
        query_id: str,
        query_start: int,
        query_end: int,
        query_length: int,
        query_strand: str,
        score: float | None = None,
        ref_name: str | None = None,
        query_name: str | None = None,
        attr: dict[str, str | float | int] | None = None
    ):
        if ref_strand == query_strand:
            ref_strand = query_strand = "+"

        if attr is None:
            self.attr: dict[str, str | float | int] = dict()
        else:
            self.attr = attr

        assert ref_strand in ("+", "-")
        assert query_strand in ("+", "-")

        self.score = None if score is None else float(score)
        
        self.ref = Region(
            ref_id,
            ref_start,
            ref_end,
            ref_length,
            ref_strand,
            ref_name
        )

        self.query = Region(
            query_id,
            query_start,
            query_end,
            query_length,
            query_strand,
            query_name
        )

    @classmethod
    def from_pafrecord(
        cls,
        pr: PafRecord,
        ref_name: str | None,
        query_name: str | None
    ) -> "Alignment":
        attr: dict[str, str | float | int] = dict()
        attr["mapq"] = int(pr.mapq)
        attr["mlen"] = int(pr.mlen)
        attr["blen"] = int(pr.blen)

        for tag in pr.tags.values():
            if tag.type == "f":
                attr[tag.tag] = float(tag.value)
            elif tag.type == "i":
                attr[tag.tag] = int(tag.value)
            elif str(tag.type) == "A":
                assert tag.value in ("P", "S", "I", "i")
                attr[tag.tag] = str(tag.value)
            elif tag.type == "Z":
                attr[tag.tag] = str(tag.value)
            else:
                raise ValueError(f"Received unexpected tag type {tag}")
        
        return cls(
            ref_id=pr.tname,
            ref_start=pr.tstart,
            ref_end=pr.tend,
            ref_length=pr.tlen,
            ref_strand="+",
            query_id=pr.qname,
            query_start=pr.qstart,
            query_end=pr.qend,
            query_length=pr.qlen,
            query_strand=str(pr.strand),
            score=pr.mapq,
            attr=attr,
            ref_name=ref_name,
            query_name=query_name
        )

    @classmethod
    def from_paffile(
        cls,
        file: PathLike,
        ref_name: str | None,
        query_name: str | None
    ) -> Iterator["Alignment"]: 
        with PafFile(file) as handle:
            for pr in handle:
                yield cls.from_pafrecord(pr, ref_name=ref_name, query_name=query_name)
        return

    def switch_ref(self) -> "Alignment":
        from copy import copy
        new = copy(self)
        new.query = self.ref
        new.ref = self.query
        if (self.query.strand == "-") and (self.ref.strand == "+"):
            new.ref.update(strand="+")
            new.query.update(strand="-")
        elif (self.ref.strand == "-") and (self.query.strand == "+"):
            new.ref.update(strand="+")
            new.query.update(strand="-")
        elif (self.ref.strand == "-") and (self.query.strand == "-"):
            new.ref.update(strand="-")
            new.query.update(strand="-")

        return new

    def __get_strand_int(self) -> int:
        if self.query.strand == self.ref.strand:
            return 1
        else:
            return -1
    
    def __repr__(self) -> str:
        return f"{self.__class__.__name__}(ref={self.ref}, query={self.query}, score={self.score}, attr={self.attr})"
    
    def plot(
        self,
        xoffset: tuple[int, bool] = (0, False), # Offset, flip
        yoffset: tuple[int, bool] = (0, False),
        ref_on_x: bool = False,
        color_value = None,
        cmap = None,
        linestyle: LINESTYLES | None = "solid",
        linewidth: float = 1.0,
        solid_capstyle: CAPSTYLES | None = "butt",
        dash_capstyle: CAPSTYLES | None = "butt",
        gapcolor: str | None = None,
        marker: MARKERS | None = "o",
        markeredgewidth: float | None = None,
        markersize: float | None = None,
        markerfacecolor = None,
        markerfacecoloralt = None,
        markeredgecolor = None,
        antialiased: bool | None = None,
        zorder: float | None = None,
        gid: str | None = None,
        snap: bool | None = None,
        sketch_params: float | tuple[float | None, float | None, float | None] | None = None, # (scale: float, length: float, randomness: float)
        ax: Axes | None = None
    ) -> mpl.lines.Line2D:
        if ax is None:
            fig, ax = plt.subplots()
    
        if color_value == "mapq":
            score = self.attr.get("mapq", 255)
            assert isinstance(score, int)
            cmap = plt.get_cmap("viridis")
            cval = cmap(score)
        else:
            cval = None

        # TODO, implement offset to get all alignments in one axis.
        strand = self.__get_strand_int()
        xoffset_, xflip = xoffset
        yoffset_, yflip = yoffset

        if ref_on_x:
            xcoords = [self.ref.start, self.ref.end][::strand]
            xlength = self.ref.length
            ycoords = [self.query.start, self.query.end]
            ylength = self.query.length
        else:
            xcoords = [self.query.start, self.query.end][::strand]
            xlength = self.query.length
            ycoords = [self.ref.start, self.ref.end]
            ylength = self.ref.length

        if xflip:
            xcoords = [xlength - x + xoffset_ for x in xcoords]
        else:
            xcoords = [x + xoffset_ for x in xcoords]

        if yflip:
            ycoords = [ylength - y + yoffset_ for y in ycoords]
        else:
            ycoords = [y + yoffset_ for y in ycoords]
        
        line = ax.plot(
            xcoords,
            ycoords,
            color=cval,
            linestyle=linestyle,
            linewidth=linewidth,
            solid_capstyle=solid_capstyle,
            dash_capstyle=dash_capstyle,
            gapcolor=gapcolor,
            marker=marker,
            markeredgewidth=markeredgewidth,
            markersize=markersize,
            markerfacecolor=markerfacecolor,
            markerfacecoloralt=markerfacecoloralt,
            markeredgecolor=markeredgecolor,
            antialiased=antialiased,
            zorder=zorder,
            gid=gid,
            snap=snap,
            sketch_params=sketch_params
        )
        return line
```

```python
class MultiAlignment(object):

    def __init__(
        self,
        alignments: Iterable[Alignment],
        ref_name: str | None = None,
        query_name: str | None = None,
        ref_fasta: Mapping[str, SeqRecord] | None = None,
        query_fasta: Mapping[str, SeqRecord] | None = None,
        ref_lengths: Mapping[str, int] | None = None,
        query_lengths: Mapping[str, int] | None = None,
        force_fasta_lengths: bool = False,
        force_alignment_lengths: bool = False,
        ignore_name_conflics: bool = False
    ):
        """
        TODO: add options for genomic features.
        """

        from collections import defaultdict
        self.alignments: defaultdict[(str, str), list[Alignment]] = defaultdict(list)

        self.ref_name = ref_name
        self.query_name = query_name

        self.ref_fasta: dict[str, SeqRecord] = {} if ref_fasta is None else dict(ref_fasta)
        self.query_fasta: dict[str, SeqRecord] = {} if query_fasta is None else dict(query_fasta)

        self.ref_lengths: dict[str, int] = {} if ref_lengths is None else dict(ref_lengths)
        self.query_lengths: dict[str, int] = {} if query_lengths is None else dict(query_lengths)

        self.add_lengths_from_seqs(force=force_fasta_lengths)
        self.add_alignments(alignments, force_lengths=force_alignment_lengths, ignore_name_conflics=ignore_name_conflics)
        return

    @classmethod
    def from_paffile(
        cls,
        file: PathLike,
        ref_name: str | None = None,
        query_name: str | None = None,
        ref_fasta: Mapping[str, SeqRecord] | None = None,
        query_fasta: Mapping[str, SeqRecord] | None = None,
        ref_lengths: Mapping[str, int] | None = None,
        query_lengths: Mapping[str, int] | None = None,
        force_fasta_lengths: bool = False,
        force_alignment_lengths: bool = False,
        ignore_name_conflics: bool = False
    ) -> "MultiAlignment":
        alignments = Alignment.from_paffile(file, ref_name=ref_name, query_name=query_name)

        return cls(
            alignments=alignments,
            ref_name=ref_name,
            query_name=query_name,
            ref_fasta=ref_fasta,
            query_fasta=query_fasta,
            ref_lengths=ref_lengths,
            query_lengths=query_lengths,
            force_fasta_lengths=force_fasta_lengths,
            force_alignment_lengths=force_alignment_lengths,
            ignore_name_conflics=ignore_name_conflics
        )

    def add_alignments(self, alignments: Iterable[Alignment], force_lengths: bool = False, ignore_name_conflics: bool = False):
        for alignment in alignments:
            if (self.ref_name is None) and (alignment.ref.name is not None):
                self.ref_name = alignment.ref.name
            elif (self.ref_name is not None) and (alignment.ref.name is None):
                alignment = alignment.update(name=self.ref_name)
            elif (
                (self.ref_name is not None) and
                (alignment.ref.name is not None) and
                (self.ref_name != alignment.ref.name) and
                (not ignore_name_conflics)
            ):
                raise ValueError(
                    f"We've previously found a name for ref ({self.ref_name}), "
                    f"but we've encountered a new ref name in the alignments ({alignment.ref.name})."
                )

            if (self.query_name is None) and (alignment.query.name is not None):
                self.query_name = alignment.query.name
            elif (self.query_name is not None) and (alignment.query.name is None):
                alignment = alignment.update(name=self.query_name)
            elif (
                (self.query_name is not None) and
                (alignment.query.name is not None) and
                (self.query_name != alignment.query.name) and
                (not ignore_name_conflics)
            ):
                raise ValueError(
                    f"We've previously found a name for query ({self.query_name}), "
                    f"but we've encountered a new query name in the alignments ({alignment.query.name})."
                )

            self.alignments[(alignment.ref.id, alignment.query.id)].append(alignment)

        self.add_lengths_from_alignments(force=force_lengths)
        return

    def add_lengths_from_seqs(self, force: bool = False):
        for seqid, seq in self.ref_fasta.items():
            seq_length = len(seq)
            if seqid in self.ref_lengths and not force:
                if self.ref_lengths[seqid] != seq_length:
                    raise ValueError(
                        f"You have previously provided a length for reference seq {seqid} ({self.ref_lengths[seqid]}), "
                        f"but it doesn't match the length in the provided fasta file {seq_length}."
                    )
            else:
                self.ref_lengths[seqid] = seq_length

        for seqid, seq in self.query_fasta.items():
            seq_length = len(seq)
            if seqid in self.query_lengths and not force:
                if self.query_lengths[seqid] != seq_length:
                    raise ValueError(
                        f"You have previously provided a length for query seq {seqid} ({self.query_lengths[seqid]}), "
                        f"but it doesn't match the length in the provided fasta file {seq_length}."
                    )
            else:
                self.query_lengths[seqid] = seq_length
        return

    def add_lengths_from_alignments(self, force: bool = False):
        ref_lengths = dict()
        query_lengths = dict()
    
        for alignments in self.alignments.values():
            for alignment in alignments:
                ref = alignment.ref
                query = alignment.query
                if ref.id in self.ref_lengths and not force:
                    if ref.length != self.ref_lengths[ref.id]:
                        raise ValueError(
                            f"You have previously provided a length for reference seq {ref.id} ({self.ref_lengths[ref.id]}), "
                            f"but it doesn't match the length in the provided alignment file {ref.length}."
                        )
                else:
                    self.ref_lengths[ref.id] = ref.length

                if query.id in self.query_lengths and not force:
                    if query.length != self.query_lengths[query.id]:
                        raise ValueError(
                            f"You have previously provided a length for query seq {query.id} ({self.query_lengths[query.id]}), "
                            f"but it doesn't match the length in the provided alignment file {query.length}."
                        )
                else:
                    self.query_lengths[query.id] = query.length
        return

    def find_order_by_length(self):
        ref_ids = sorted(self.ref_lengths.keys(), key=lambda k: self.ref_lengths[k], reverse=True)
        query_ids = sorted(self.query_lengths.keys(), key=lambda k: self.query_lengths[k], reverse=True)
        return ref_ids, query_ids

    def find_offsets(
        self,
        ref_order: Iterable[str | tuple[str, bool]],
        query_order: Iterable[str | tuple[str, bool]],
        gap: int = 0
    ) -> tuple[list[tuple[str, int, int, bool]], list[tuple[str, int, int, bool]]]:
        # Seqid, start, end, flip
        # I include the end coord in case we have a gap > 0.
        # It also gives us the final length
        ref_offsets: list[tuple[str, int, int, bool]] = []
        ref_offset = 0
        for ref in ref_order:
            if isinstance(ref, tuple):
                ref_id, ref_flip = ref
                assert isinstance(ref_id, str)
                assert isinstance(ref_flip, bool)
            else:
                assert isinstance(ref, str)
                ref_id = ref
                ref_flip = False

            if ref_id not in self.ref_lengths:
                raise ValueError(
                    f"Your reference order included seqid {ref_id}, but there are no alignments or sequence length for this. "
                    f"Please add some alignments including {ref_id}, or directly add the length to obj.ref_lengths"
                )

            ref_length = self.ref_lengths[ref_id]
            ref_offsets.append((ref_id, ref_offset, ref_offset + ref_length, ref_flip))
            ref_offset += ref_length + gap

        query_offsets: list[tuple[str, int, int, bool]] = []
        query_offset = 0
        for query in query_order:
            if isinstance(query, tuple):
                query_id, query_flip = query
            else:
                query_id = query
                query_flip = False

            if query_id not in self.query_lengths:
                raise ValueError(
                    f"Your query order included seqid {query_id}, but there are no alignments or sequence length for this. "
                    f"Please add some alignments including {query_id}, or directly add the length to obj.query_lengths"
                )

            query_length = self.query_lengths[query_id]
            query_offsets.append((query_id, query_offset, query_offset + query_length, query_flip))
            query_offset += query_length + gap

        return ref_offsets, query_offsets

    def plot_single_axis(
        self,
        color_value = "mapq",
        ref_on_x: bool = False,
        margin: int = 0,
        ax: Axes = None
    ):
        if ax is None:
            fig, ax = plt.subplots()
            ax.set_aspect(1.0)

        ref_order, query_order = self.find_order_by_length()
        ref_offsets, query_offsets = self.find_offsets(ref_order, query_order)

        if ref_on_x:
            x_offsets = ref_offsets
            y_offsets = query_offsets
        else:
            x_offsets = query_offsets
            y_offsets = ref_offsets

        x_ids = [t[0] for t in x_offsets]
        x_min = min((t[1] for t in x_offsets))
        x_max = max((t[2] for t in x_offsets))
        x_starts = [t[1] for t in x_offsets]
        x_mids = [t[1] + abs(t[1] - t[2]) / 2 for t in x_offsets]

        y_ids = [t[0] for t in y_offsets]
        y_min = min((t[1] for t in y_offsets))
        y_max = max((t[2] for t in y_offsets))
        y_mids = [t[1] + abs(t[1] - t[2]) / 2 for t in y_offsets]
        y_starts = [t[1] for t in y_offsets]

        for x_id, x_start, x_end, x_flip in x_offsets:
            for y_id, y_start, y_end, y_flip in y_offsets:
                if ref_on_x:
                    alignments = self.alignments.get((x_id, y_id), [])
                else:
                    alignments = self.alignments.get((y_id, x_id), [])

                for a in alignments:
                    a.plot(
                        color_value="mapq",
                        ax=ax,
                        marker=".",
                        ref_on_x=ref_on_x,
                        xoffset=(x_start, x_flip),
                        yoffset=(y_start, y_flip)
                    )

        ax.set_xlim(x_min - margin, x_max + margin)
        ax.set_ylim(y_min - margin, y_max + margin)
        
        ax.set_xticks(x_starts, labels=x_ids)
        ax.set_yticks(y_starts, labels=y_ids)

        ax.grid()

        return ax
```
